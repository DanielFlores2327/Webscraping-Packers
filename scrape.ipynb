```python
url = 'https://www.pro-football-reference.com/teams/gnb/2023/gamelog/'
print(url)
```

```python
off_df = pd.read_html(url, header=1, attrs={'id':'gamelog' + str(2023)})[0]
```

```python
def_df = pd.read_html(url, header=1, attrs={'id':'gamelog_opp' + str(2023)})[0]
```

```python
team_df = pd.concat([off_df, def_df], axis=1)
```

```python
off_df.to_csv('offdata.csv', index=False)
```

```python
team_df.to_csv('first_data.csv', index=False)
```

```python
numeric_df = off_df.select_dtypes(include=[np.number])
numeric_df.corr()
```

So, for the packers at least, touchdowns and extra points attempted correlated the most to points but that is sort of trivial.

Furthermore, rushing attempts correlated to winning more than passing attempts but that is more of a lagging indicator because teams that are winning/scored points are gonna rush more because of already winning.

However, Field Goals attempted correlated negatively to points which is quite the surprise considering that attempted FGs leads to more points however this may lead to the insight that field goals don't produce that many points and for a good offense like the packers they should go for it on fourth down instead more often that a shot at 3 points especially since anders carlson was so bad.
